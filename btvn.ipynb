## 1. Gộp dữ liệu theo 3 bảng để tạo ra bảng phân tích tổng hợp df_all:

# YÊU CẦU 3

## 1. Tạo thêm các cột:
– DoanhThu = SoLuong × GiaBan

– Thang = tháng của cột Ngay

```python
df_all['DoanhThu'] = df_all['SoLuong'] * df_all['GiaBan']
```

```python
df_all['Thang'] = df_all['Ngay'].dt.month
```

```python
df_all.head()
```

## 2. Phân tích tổng quan:
– Tổng doanh thu toàn hệ thống

– Doanh thu theo từng cửa hàng

– Doanh thu theo từng tỉnh/thành

– Doanh thu theo danh mục sản phẩm

```python
# Tong doanh thu toan he thong
dt_toanhethong = df_all['DoanhThu'].sum()
print(format(dt_toanhethong,',d'))
```

```python
df_all.info()
```

```python
# Doanh thu theo tung cua hang
dt_cuahang = df_all.groupby('MaCuaHang')['DoanhThu'].sum().reset_index()
print(dt_cuahang)
```

```python
# Doanh thu theo tung tinh/thanh
dt_tinhthanh = df_all.groupby('TinhThanh')['DoanhThu'].sum().reset_index()
print(dt_tinhthanh)
```

```python
# Doanh thu theo danh muc san pham
dt_danhmuc = df_all.groupby('DanhMuc')['DoanhThu'].sum().reset_index()
print(dt_danhmuc)
```

```python
df_hanghoa['DanhMuc'].unique()
```

## 3. Sản phẩm & Cửa hàng:
– Top 10 sản phẩm bán chạy nhất (theo SoLuong)

– Top 10 sản phẩm có doanh thu cao nhất

– Top 5 cửa hàng doanh thu cao nhất

– Top 5 cửa hàng có lượng bán thấp nhất (theo SoLuong)

```python
# 10 san pham ban chay theo so luong
top_sanpham = df_all.groupby('TenHang')['SoLuong'].sum().sort_values(ascending=False).head(10).reset_index()
print(top_sanpham)
```

```python
# 10 san pham co doanh thu cao nhat
topdt_sanpham = df_all.groupby('TenHang')['DoanhThu'].sum().sort_values(ascending=False).head(10)
print(topdt_sanpham)
```

```python
# 5 cua hang co doanh thu cao nhat
top_cuahang = df_all.groupby('TenCuaHang')['DoanhThu'].sum().sort_values(ascending=False).head(5)
print(top_cuahang)
```

```python
# 5 cua hang co luong ban thap nhat theo so luong
top_cuahang_lowest = df_all.groupby('TenCuaHang')['SoLuong'].sum().sort_values(ascending=True).head(5)
print(top_cuahang_lowest)
```

## 4. Phân tích theo vị trí cửa hàng:
– So sánh doanh thu TB của các loại vị trí: Khu dân cư, Văn phòng, Trường học, Khu công nghiệp, Trung tâm thương mại.

– Xác định loại vị trí có: Doanh thu trung bình cao nhất, Số giao dịch nhiều nhất.

```python
# So sanh doanh thu TB cua tung vi tri
dt_khuvuc = df_all.groupby('ViTri')['DoanhThu'].mean().reset_index()
print(dt_khuvuc)
```

```python
# Xac dinh loai vi tri co doanh thu TB cao nhat, so giao dich nhieu nhat
## Doanh thu TB cao nhat
dttb_top_khuvuc = df_all.groupby('ViTri')['DoanhThu'].mean().sort_values(ascending=False).head(1)
print(dttb_top_khuvuc)
## So giao dich nhieu nhat
sogd_top_khuvuc = df_all.groupby('ViTri').size().sort_values(ascending=False).head(1)
print(sogd_top_khuvuc)
```

```python
top_khuvuc = df_all.groupby('ViTri')[['DoanhThu','MaGD']].agg({'DoanhThu':'mean','MaGD':'count'})
print(top_khuvuc)
```
